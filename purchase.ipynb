```python
#Top Senders

temp = data
temp['Count'] = data.groupby('SN')['SN'].transform('count')
tempp = temp[['SN', 'Price']].groupby('SN').sum()
temppp = temp[['SN', 'Count']]
s = pd.merge(tempp, temppp, on='SN')
s['Avg'] = s['Price'] / s['Count']
ans = s.drop_duplicates().sort_values(by='Count', ascending = 0)
ans.head()
```

```python
#Most Popular Items

temp = data
temp['Count'] = data.groupby('Item ID')['Item ID'].transform('count')
temp = temp[['Item ID', 'Item Name', 'Count', 'Price']].drop_duplicates()
temp = temp.copy()
temp['Total Purchase value'] = temp['Price'] * temp['Count']
ans = temp.sort_values(by='Count', ascending = 0)
ans.head()
```

```python
#Most Profitable Items

temp = data
temp['Count'] = data.groupby('Item ID')['Item ID'].transform('count')
temp = temp[['Item ID', 'Item Name', 'Count', 'Price']].drop_duplicates()
temp = temp.copy()
temp['Total Purchase value'] = temp['Price'] * temp['Count']
ans = temp.sort_values(by='Total Purchase value', ascending = 0)
ans.head()
```
